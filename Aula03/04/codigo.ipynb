```python
#chama as bibliotecas
import matplotlib.pyplot as plt 
import numpy as np 
```

```python

Dados_x = np.linspace(0, 10, 10) # gera 10 valores de 10 em 10
Dados_y = Dados_x * Dados_x
#linspace gera valores espaçados de 10 em 10
```

```python
plt.figure(figsize = (10,8), facecolor = "gray" ) #plt.figure

#cria figura
#figsize tamanho da figura
ax = plt.axes()
ax.set_facecolor("white") #Cor do fundo do gráfico
plt.plot(Dados_x,Dados_y,color = "red", marker = 'o', linestyle= 'dashed', markeredgecolor = "black")

#Plota o gráfico x e y
plt.xlabel("Eixo X", fontsize = 16)
plt.ylabel("Eixo Y", fontsize = 16)
plt.title("Gráfico simples com Matplotlib", fontsize = 16)
```

```python
plt.rcParams.update({'font.size':15}) #Função para ajustar a fonte

#figura
plt.figure(figsize = (12, 12)) #Define o tamanho da figura
linhas = 2 #Cria duas linhas
colunas = 2 #Cria duas colunas

#Subplot para plotar as figuras
plt.subplot(linhas, colunas, 1) #plota grafico na primeira posição
plt.plot(Dados_x, "r--", label = "Dados_x")
plt.legend(loc= 'lower right') #Cria legenda para o gráfico
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title("Gráfico 1")

#2° Gráfixo
plt.subplot (linhas, colunas, 2)
plt.plot(Dados_x, Dados_y, "b--", label = "Dados_x \n e y")

#Plota gráfico alterando a cor da linha
plt.title("Gráfico 2") #Titulo do gráfico
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend
```

```python
#Grafico 3
plt.subplot(linhas,colunas, 3)
plt.plot(Dados_x * Dados_x)
```

```python
#Grafico de barras

Figura = plt.figure () #Cria uma variável chamada figura
Grafico = Figura.add_axes([0, 0, 0.7, 0.7]) #Cria eixos para os graficos 

#Cria rotulos
Rotulos = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']

#Plota o gráfico
Grafico.bar(Rotulos, Dados_x, color = 'red') #Plota grafico de barras
Grafico.set_xlabel('Rotulos')
Grafico.set_ylabel('Quantidade')
```

```python
#Criando o segundo grafico de barras

Grafico.bar(Rotulos, Dados_x, color = "red") #Dados 1
Grafico.bar(Rotulos, Dados_y * 0.1, color = "blue") #Dados 2

Figura
```
